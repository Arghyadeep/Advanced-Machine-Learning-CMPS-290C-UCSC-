```python
from sklearn.decomposition import PCA
from sklearn.externals import joblib
import matplotlib.pyplot as plt
```

```python
tsne_data = joblib.load("tsnedata.pkl")
tsne_labels = joblib.load("tsnelabel.pkl")
pca = PCA(n_components=2)
pca_result = pca.fit_transform(tsne_data)
joblib.dump(pca_result,"PCA.pkl")
print(pca_result)
```

```python
plt.scatter(pca_result[:,0],pca_result[:,1],10,tsne_labels)
plt.show()
```
